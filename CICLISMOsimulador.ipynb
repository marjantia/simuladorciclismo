```python
import random
import datetime

# Definir la distancia de la etapa en kilómetros
distance = 150

# Crear una lista de nombres de corredores
names = ['Mario', 'Luigi', 'Peach', 'Toad', 'Yoshi', 'Wario', 'Donkey Kong', 'Bowser', 'Daisy', 'Koopa']

# Definir una lista de equipos y sus habilidades correspondientes.
teams = [{'name': 'Equipo A', 'skill': 0.8}, {'name': 'Equipo B', 'skill': 0.6}, {'name': 'Equipo C', 'skill': 1.0}]

# Crear una lista vacía para almacenar los tiempos de los corredores
times = []

# Crear un diccionario vacío para almacenar los tiempos globales de los equipos
team_global_times = {'Equipo A': 0, 'Equipo B': 0, 'Equipo C': 0}

def simulate_performance(team_skill, fatigue_level):
    # Convertir la habilidad del equipo a un número (flotante)
    team_skill = float(team_skill)

    # Definir la velocidad máxima en función de la habilidad del equipo
    max_speed = 30 + 10 * team_skill  # kilómetros por hora

    # Aplicar un factor de fatiga en función del nivel de fatiga
    max_speed *= 1 - 0.5 * fatigue_level  # Reducir la velocidad máxima en función del nivel de fatiga
    speed = random.uniform(0.7, 1) * max_speed
    time = distance / speed

    # Agregar una variación aleatoria de +/- 5 minutos
    time += random.uniform(-0.08, 0.08)
    return time

# Pedir al usuario que ingrese el nivel de esfuerzo y el nivel de fatiga de cada corredor
levels_of_effort = []
fatigue_levels = []
print('Ingrese el nivel de esfuerzo y el nivel de fatiga de cada corredor (intenso, medio o leve; un valor entre 0 y 1 para la fatiga):')
for name in names:
    level_of_effort, fatigue_level = input(f'{name}: ').split()
    while level_of_effort not in ['intenso', 'medio', 'leve'] or not (0 <= float(fatigue_level) <= 1):
        print('Los valores ingresados no son válidos.')
        level_of_effort, fatigue_level = input(f'{name}: ').split()
    levels_of_effort.append(level_of_effort)
    fatigue_levels.append(float(fatigue_level))

team_skills = []  # Lista adicional para almacenar la habilidad del equipo de cada corredor
print('Ingrese el equipo de cada corredor:')
for name in names:
    team = input(f'{name}: ')
    while team not in [t['name'] for t in teams]:
        print('El equipo ingresado no es válido.')
        team = input(f'{name}: ')
    team_skill = [t['skill'] for t in teams if t['name'] == team][0]
    team_skills.append(team_skill)  # Se agrega la habilidad del equipo a la lista team_skills

# Recorrer la lista de nombres de los corredores y simular su tiempo en la etapa
for i, name in enumerate(names):
    team_skill = team_skills[i]
    fatigue_level = fatigue_levels[i]
    time = simulate_performance(team_skill, fatigue_level)
    times.append((name, time))

# Pedir al usuario que ingrese los tiempos globales de los corredores
print('Ingrese los tiempos globales de los corredores:')
for i, (name, time) in enumerate(times):
    global_time_str = input(f'{i+1}. {name}: ')
    try:
        # Convertir el tiempo global ingresado en un objeto timedelta
        global_time = datetime.datetime.strptime(global_time_str, '%H:%M:%S') - datetime.datetime(1900, 1, 1)
        # Calcular el nuevo tiempo del corredor
        new_time = (global_time + datetime.timedelta(hours=time)).total_seconds() / 3600
        # Actualizar el tiempo del corredor en la lista de tiempos
        times[i] = (name, new_time)
        # Redondear el nuevo tiempo
        new_time_rounded = round(new_time * 3600) / 3600
        # Convertir el nuevo tiempo redondeado en un objeto timedelta
        new_time_timedelta = datetime.timedelta(hours=new_time_rounded)
        # Imprimir el nuevo tiempo del corredor en formato HH:MM:SS
        print(f'{name}: {new_time_timedelta}')
        # Actualizar el tiempo global del equipo al que pertenece el corredor
        team = input(f'Equipo de {name}: ')
        while team not in [t['name'] for t in teams]:
            print('El equipo ingresado no es válido.')
            team = input(f'Equipo de {name}: ')
        team_global_times[team] += new_time
    except ValueError:
        print(f'Error: el formato de tiempo ingresado para {name} no es válido.')

# Ordenar la lista de tiempos actualizada de menor a mayor
times.sort(key=lambda x: x[1])

# Imprimir la clasificación final actualizada
print('Clasificación final actualizada:')
for i, (name, time) in enumerate(times):
    # Redondear el tiempo a HH:MM:SS exactos
    time_rounded = round(time * 3600) / 3600
    # Convertir el tiempo redondeado en un objeto timedelta
    time_timedelta = datetime.timedelta(hours=time_rounded)
    # Imprimir la clasificación con el tiempo redondeado en formato HH:MM:SS
    print(f'{i+1}. {name}: {str(time_timedelta)}')

# Ordenar y mostrar la clasificación global por equipos
sorted_team_times = sorted(team_global_times.items(), key=lambda x: x[1])
print('Clasificación global por equipos:')
for i, (team, time) in enumerate(sorted_team_times):
    time_rounded = round(time * 3600) / 3600
    time_timedelta = datetime.timedelta(hours=time_rounded)
    print(f'{i+1}. {team}: {str(time_timedelta)}')
```
